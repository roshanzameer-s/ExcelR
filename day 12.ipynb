```python
# Import necessary libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1: Define the text
text = 'data science machine learning artificial intelligence'

# Step 2: Generate the WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Step 3: Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.title("WordCloud of Given Text", fontsize=16)
plt.show()

# Step 4: Save the WordCloud as an image
wordcloud.to_file("wordcloud_output.png")
print("WordCloud saved as 'wordcloud_output.png'")
```
